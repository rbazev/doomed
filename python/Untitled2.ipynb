```python
# plt.figure(figsize=(6, 6))
count = 0
pos = {}
x = [0] * 81
y = [0] * 81
for i in range(9):
    for j in range(9):
        x[count] = i 
        y[count] = j
        pos.update({count: [i, j]})
        count += 1
n = count
# plt.plot(x, y, 'o')
# for i in range(n):
#     plt.text(x[i], y[i], str(i))
# plt.axis('off');
```

```python
g = nx.Graph()
pos = {n-1: (x[n-1], y[n-1])}
for i in range(n-1):
    pos.update({i: (x[i], y[i])})
    for j in range(i+1, n):
        if i != j:
            d = distance.euclidean((x[i], y[i]), (x[j], y[j]))
            if d < 1.415:
                g.add_edge(i, j)
plt.figure(figsize=(10,10))
nx.draw_networkx_nodes(g, pos=pos, node_color='w', edgecolors='k')
nx.draw_networkx_edges(g, pos=pos)
nx.draw_networkx_labels(g, pos=pos, labels={i:i+1 for i in pos.keys()}, )
alledges = g.edges
deg = np.array(g.degree())[:,1]
print(np.mean(deg), np.std(deg))
plt.axis('off')
plt.savefig('nodes.pdf')
```
