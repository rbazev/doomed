```python
red = sns.xkcd_rgb["pale red"]
gol = sns.xkcd_rgb["golden"]
ora = sns.xkcd_rgb["light orange"]
gre = sns.xkcd_rgb["dark grey"]
grn = sns.xkcd_rgb["medium green"] 
blu = sns.xkcd_rgb["azure"]
```

```python
data = pd.read_csv('../data/pars.csv', encoding = "ISO-8859-1")
data.loc[data['hom'],'s'] = data['s'] / 2
del data['hom']
data = data[pd.notna(data['U'])]
print('n =', len(data), 'estimates')
data
```

```python
data['U'].median(), data['s'].median(), 
```

```python
spp = data['species'].value_counts()
print('n =', len(spp), 'species')
spp
```

```python
grp = data['group'].value_counts()
grp
```

```python
ref = data['source'].value_counts()
print('n =', len(ref), 'studies')
ref
```

```python
logu = np.log10(data['U'])
logu.min(), logu.max() 
```

```python
logs = np.log10(data['s'])
logs.min(), logs.max() 
```

```python
mu = logu.median()
ms = logs.median()
10**mu, 10**ms
```

```python
Ulo = 10**-2.5
Uhi = 10**-.85
slo = 10**-1.6
shi = 10**-.85
tmp = data[(data['U'] < Ulo)]
print('Ulo =', Ulo, '->', len(tmp))
tmp = data[(data['U'] > Uhi)]
print('Uhi =', Uhi, '->', len(tmp))
tmp = data[(data['s'] < slo)]
print('slo =', slo, '->', len(tmp))
tmp = data[(data['s'] > shi)]
print('shi =', shi, '->', len(tmp))
tmp = data[(data['U'] >= Ulo) & (data['U'] <= Uhi) & (data['s'] >= slo) & (data['s'] <= shi)]
print(len(tmp))
```

```python
tmp
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_position([.16, .16, .83, .83])
grps = ['virus', 'bacteria', 'eukaryote']
symbs = ['o','s','v']
cols = [ora, grn, blu]
ax.plot(mu, ms, '+', ms=15, color=red, markeredgewidth=2)
for i in range(3):
    tmp = data[data['group']==grps[i]]
    ax.plot(np.log10(tmp['U']), np.log10(tmp['s']), 'o', label=grps[i], color=cols[i])
ax.plot(np.log10([Ulo] * 2), np.log10([slo, shi]), '-', color=red)
ax.plot(np.log10([Uhi] * 2), np.log10([slo, shi]), '-', color=red)
ax.plot(np.log10([Ulo, Uhi]), np.log10([slo] * 2), '-', color=red)
ax.plot(np.log10([Ulo, Uhi]), np.log10([shi] * 2), '-', color=red)
ax.set_xlabel('Deleterious mutation rate $(u)$')
ax.set_xlim(-4.3, .3)
ax.set_xticks([-4, -2, 0])
ax.set_xticklabels([r'$10^{-4}$', r'$10^{-2}$', 1])
ax.set_ylabel('Selection coefficient $(s)$')
ax.set_ylim(-2.8, -.5)
ax.set_yticks([-2, -1])
ax.set_yticklabels([r'$10^{-2}$', r'$10^{-1}$'])
ax.legend(fontsize=14)
plt.savefig('../figures/pars.pdf')
```
